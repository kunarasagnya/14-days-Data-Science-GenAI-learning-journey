#  Day 4 – RAG (Retrieval-Augmented Generation) Theory


##  What is RAG?
RAG (Retrieval-Augmented Generation) allows LLMs to retrieve relevant information
from an external knowledge base and use it to generate accurate and grounded answers.



##  Why RAG?
- LLMs have training cut-off dates  
- They hallucinate missing facts  
- They cannot recall your documents unless retrieved  
- RAG gives them access to your custom data  



##  RAG Workflow (Step-by-Step)
1. **User Query**
2. **Embedding Model converts query to vector**
3. **Vector DB retrieves top-k similar chunks**
4. **Relevant chunks returned to system**
5. **Query + Context sent to LLM**
6. **LLM generates grounded answer**



##  RAG Components
- Chunking  
- Embeddings  
- Vector DB (FAISS, Chroma, Pinecone)  
- Retriever  
- LLM

##  Limitations of RAG
- Bad chunking → bad retrieval  
- Needs quality embeddings  
- Retrieval may return irrelevant chunks  
- Not suitable for pure reasoning tasks  



##  When NOT to use RAG
- Coding tasks  
- Math-heavy reasoning  
- When answers do not depend on external documents
